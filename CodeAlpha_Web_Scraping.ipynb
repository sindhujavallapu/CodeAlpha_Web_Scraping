## Web Scraping of Books Data Using Python and BeautifulSoup

### Objective
Extract book information from a website and create a dataset for analysis.

### Tools Used
- Python
- Requests
- BeautifulSoup
- Pandas
- Google Colab

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com/"
response = requests.get(url)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
books = soup.find_all("article", class_="product_pod")

titles = []

for book in books:
    title = book.h3.a["title"]
    titles.append(title)

titles
```

```python
prices = []

for book in books:
    price = book.find("p", class_="price_color").text
    prices.append(price)

prices
```

```python
df = pd.DataFrame({
    "Book Title": titles,
    "Price": prices
})

df.head()
```

```python
df.to_csv("Books_Data.csv", index=False)
```

```python
from google.colab import files
files.download("Books_Data.csv")
```

## Conclusion

Successfully scraped book titles and prices from a website using BeautifulSoup and created a structured dataset using Pandas.
